```python
from pygments import highlight
from pygments.lexers import PythonLexer
from pygments.formatters import HtmlFormatter
from pygments.styles import get_all_styles

for i in range(168):
    fname = 'codes/' + str(i) + '.py'
    with open(fname) as f:
        content = f.read()
    html = highlight(content, PythonLexer(), HtmlFormatter(linenos=True, cssclass="source", noclasses=True))
    print(html)
    fname = 'htmls/' + str(i) + '.html'
    with open(fname, 'w+') as f:
        content = f.write(html)
```
